```python
cm = df.corr()
mask = np.triu(np.ones_like(cm, dtype=bool))

# Set up the figure size
plt.figure(figsize=(12, 8))

# Create the heatmap with the mask
sns.heatmap(cm, annot=True, cmap='mako', fmt=".2f", mask=mask)

# Show the plot
plt.title('Correlation Heatmap for Walmart Data')
plt.show()
```

```python
plt.figure(figsize=(15, 10))

# Create box plots for all columns
sns.boxplot(data=df, orient='h', palette='Set2')  # 'orient' is set to 'h' for horizontal orientation

# Set plot title
plt.title('Box Plots for Outlier Detection in All Columns')

# Show the plot
plt.show()
```

```python
outlier_threshold = 1.5

# Iterate through each column in the DataFrame
for column in df.columns:
    # Calculate the IQR for the current column
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Identify outliers based on the threshold
    outliers = (df[column] < Q1 - outlier_threshold * IQR) | (df[column] > Q3 + outlier_threshold * IQR)
    
    # Calculate the percentage of outliers
    percentage_outliers = outliers.sum() / len(df) * 100
    
    # Print column name and percentage of outliers if they exist
    if percentage_outliers > 0:
        print(f"Column '{column}' has {percentage_outliers:.2f}% outliers.")
```

```python
Q3 = df['Weekly_Sales'].quantile(0.75)
Q1 = df['Weekly_Sales'].quantile(0.25)
IQR = Q3-Q1
upperLimit = Q3+(1.5*IQR)
lowerLimit = Q1-(1.5*IQR)
filt = ((df['Weekly_Sales']<= upperLimit) & (df['Weekly_Sales']>= lowerLimit))
df = df[filt]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Assuming you have a DataFrame 'df' with your predictor variables


# Create a DataFrame with only the predictor variables
predictors = df[['Weekly_Sales', 'Holiday_Flag', 'Temperature', 'Unemployment', 'Fuel_Price', 'CPI', 'Store']]


# Calculate VIF for each predictor
vif_data = pd.DataFrame()
vif_data["Variable"] = predictors.columns
vif_data["VIF"] = [variance_inflation_factor(predictors.values, i) for i in range(predictors.shape[1])]

# Display the VIF values
print(vif_data)
```
